```python
import numpy as np
import re
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
# %matplotlib notebook

#-?\d+
# -? = optional minus sign
# \d = decimal digit, [0-9]
# +  = repeating decimal
# data extraction
data_txt = open('input.txt','r').read().splitlines()
positionX, positionY, velocityX, velocityY = np.array(
    [re.findall('-?\d+', entry) for entry in data_txt], int).T

# skip ahead
positionX += velocityX * 10000
positionY += velocityY * 10000

# initialization figure
fig = plt.figure(figsize=(8, 6))
ax = plt.axes(xlim=(100, 180), ylim=(140, 180))
ax.invert_yaxis()
line, = ax.plot([], [], 'o')

def init():
    line.set_data([], [])
    return line,

def animate(i, positionX, positionY):
    ax.set_xlabel(i)
    line.set_data(positionX,positionY)
    positionX += velocityX
    positionY += velocityY
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, fargs=(positionX, positionY),
                               frames=38, interval=300, blit=True)
# anim.save('basic_animation.mp4')
HTML(anim.to_html5_video())
# 10036 seconds
```

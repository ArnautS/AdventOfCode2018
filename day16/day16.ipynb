```python
import numpy as np
import re

def addr(A,B,C):
    reg[C] = reg[A] + reg[B]
    
def addi(A,B,C):
    reg[C] = reg[A] + B
    
def mulr(A,B,C):
    reg[C] = reg[A] * reg[B]

def muli(A,B,C):
    reg[C] = reg[A] * B
    
def banr(A,B,C):
    reg[C] = reg[A] & reg[B]
    
def bani(A,B,C):
    reg[C] = reg[A] & B
    
def borr(A,B,C):
    reg[C] = reg[A] | reg[B]
    
def bori(A,B,C):
    reg[C] = reg[A] | B
    
def setr(A,B,C):
    reg[C] = reg[A]
    
def seti(A,B,C):
    reg[C] = A
    
def gtir(A,B,C):
    reg[C] = 1 if A > reg[B] else 0
    
def gtri(A,B,C):
    reg[C] = 1 if reg[A] > B else 0
    
def gtrr(A,B,C):
    reg[C] = 1 if reg[A] > reg[B] else 0
    
def eqir(A,B,C):
    reg[C] = 1 if A == reg[B] else 0
    
def eqri(A,B,C):
    reg[C] = 1 if reg[A] == B else 0
    
def eqrr(A,B,C):
    reg[C] = 1 if reg[A] == reg[B] else 0

operations = [addr, addi, mulr, muli, banr, bani, borr, bori, setr, seti, gtir,
              gtri, gtrr, eqir, eqri, eqrr]

opDict = np.zeros((16,16), int)

reg = np.zeros(4, int)
    
text = re.split('\n\n\n',open('input.txt','r').read())
part1 = re.split('\n\n',text[0])
data1 = np.array([np.array([np.array([int(x) for x in re.findall('\d+',line)]) for line in y.splitlines()]) for y in part1])
res = 0
for sample in data1:
    reg = np.copy(sample[0])
    instruction = sample[1][1:]
    behavesLikeOpcode = 0
    for i, op in enumerate(operations):
        op(*instruction)
        if (reg == sample[2]).all():
            behavesLikeOpcode += 1
            opDict[i][sample[1][0]] += 1
        reg = np.copy(sample[0])
    if behavesLikeOpcode >= 3:
        res += 1

print(res)

indices = np.arange(16)
operationsOrdered = np.copy(operations)
for j in range(16):
    for i, op in enumerate(opDict.T):
        matches = indices[op>0]
        if len(matches) == 1:
#             print(operations[matches[0]],'=',i)
            operationsOrdered[i] = operations[matches[0]]
            opDict[matches[0],:] = 0

data2 = [[int(x) for x in re.findall('\d+',line)]for line in text[1][1:].splitlines()]
reg = np.zeros(4, int)
for instruction in data2:
    operationsOrdered[instruction[0]](*instruction[1:])
print(reg)
    
```

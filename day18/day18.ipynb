```python
import numpy as np
from itertools import product
from IPython.display import clear_output
import time

data = open('input.txt','r').read().splitlines()
acres = np.array([[x for x in line] for line in data])
size = acres.shape[0]
values = []
looped = 0

def printer(data):
    for line in data:
        print(''.join(line))

def neighbours(cell):
    for c in product(*(range(n-1, n+2) for n in cell)):
        if c != cell and all(0 <= n < size for n in c):
            yield oldAcres[c]
    
# acres = np.empty((size,size),str)
#     acres[:] = ' '
    
for i in range(2000):
    oldAcres = np.copy(acres)
    for index, acre in np.ndenumerate(oldAcres):
        adjacent = ''.join(list(neighbours(index)))
        if acre == '.' and adjacent.count('|') >= 3:
            acres[index] = '|'
        elif acre == '|' and adjacent.count('#') >= 3:
            acres[index] = '#'
        elif acre == '#' and not ('#' in adjacent and '|' in adjacent):
            acres[index] = '.'
#     if i > 100:
#         clear_output()
#         printer(acres)
#         time.sleep(0.5)
        
    unique, counts = np.unique(acres, return_counts=True)
    counted = dict(zip(unique, counts))
    value = counted['#']*counted['|']
    if value in values:
        values.append(value)
        looped += 1
        print('looped at', i)
        if looped == 40:
            break
    else:
        values.append(value)

unique, counts = np.unique(acres, return_counts=True)
counted = dict(zip(unique, counts))
print(counted['#']*counted['|'])
printer(acres)

# too low 57572
```
